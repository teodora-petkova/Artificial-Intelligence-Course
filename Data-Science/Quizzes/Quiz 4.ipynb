```python
sofia = {"lat": 42.6977, "lng": 23.3219}
m = Basemap(
    projection = "merc",
    llcrnrlat = 41,
    llcrnrlon = 22,
    urcrnrlat = 44.5,
    urcrnrlon = 29,
    resolution = "l")
m.drawcoastlines()
m.drawcountries()
x, y = m(sofia["lng"], sofia["lat"])
m.plot(x, y, "o", markersize = 2, color = "red")
plt.show()
```

```python
g = nx.florentine_families_graph()
nx.draw(g, with_labels = True)
plt.show()
```

```python
shortest_path_Pazzi_Peruzzi = nx.shortest_path(g, "Pazzi", "Peruzzi")
print("Shortest path between Pazzi & Peruzzi:", shortest_path_Pazzi_Peruzzi)
print("Shortest path between Pazzi & Peruzzi length:", len(shortest_path_Pazzi_Peruzzi))
```

```python
families_with_centrality = nx.betweenness_centrality(g)
sorted_by_centrality = sorted(families_with_centrality.items(),
                              key = lambda item: item[1])[::-1]
print("The most important family:", sorted_by_centrality[0])
```
