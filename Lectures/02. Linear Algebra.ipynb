# Linear Algebra

## Vectors

Дефиниция на вектор:
- a pointed segment in space; насочена отсечка с начало; описва се с посока и големина
- в програмирането: едномерен вектор от числа (вектор стълб)

Свободен вектор - вектор, който може да местим навсякъде, инвариантен спрямо транслация

Примери: всяка една сила, гравитацията, скоростта

Можем да си представим, че векторите могат да образуват нещо като поле, но не е точно поле
- събиране на вектори
- умножението на вектор със скалар/число, но не можем да умножаваме вектор с вектор

Друга перспектива - трансформация, tensors

Всеки вектор може да се движи свободно, но може да прикачим координатна система към него.
За улеснение може да вземем началото на вектора за начало на координатната система и става радиус-вектор и правим декартова координатна система (правоъгълна)

Операциите с вектори:
- събиране - последователно (((a+b)+c)+d) - the order does not matter
- умножение с число/скалар
(scalar - scale/скалиране на размера на векторите)

Видове вектори:
- колинеарни вектори = успоредни (т.е. правите, на които лежат са успоредни)
- еднопосочни колинеарни 
- разнопосочни колинеарни 

Умножение между вектори:

1.) Скаларно произведение $ a. b = \mid a \mid \mid b \mid cos(\phi) $

2.) Векторно произведение $ a \times b $

А какво образуват векторите, ако не е поле?

### Vector space - линейно векторно пространство - 

 - елементи: вектори и поле $F$ - real or complex numbers
 - операции: събиране на вектори + умножение на вектор с число

*Аксиоми*:
- нулев вектор
- единичен вектор
- дистрибутивност
- асоциативност в ляво и десно
- комукативност

Примери: 
- пространство на координатите
- полиномите x + real coefficients
- пространство на функциите - събиране на функции

пръстен - затворено множество относно събиране и умножение

### Linear combinations - умножение на вектор с числа
- линейна обвивка - linear hull - множеството от всички възможните линейните комбинации => span

### Matrix
- таблица от числа - row x colums / редове и стълбове

### Linear transformations
 a mapping/function between two vector spaces
- запазва успоредните прави (shear - разтегляне)
- поне една точка трябва да остане, където е била

Примери:
- ротация
- shear (завърта и разтегля)
- distortion - no!
- translation - no!
 
проекция от двумерно върху едномерно пространство (или 30-мерно върху 3-мерно) -> намаляване на мерността

трансформация = смяна на базиса
относителните координати остават същите - само числата се променят, числа спрямо новия базис

### Transformation Matrix
- записа на двата базисни вектори по стълбове
vector = matrix * vector
