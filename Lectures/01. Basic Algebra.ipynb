```python
x = np.linspace(-2 * np.pi, 2 * np.pi, 5000) # linear space
```

```python
y1= np.sin(x)
y2= np.cos(x)
```

```python
plt.plot(x, y1, label = "sin(x)")
plt.plot(x, y2, label = "cos(x)")
plt.legend
plt.show()
```

```python
def plot_math_function(functions, x_min, x_max, num_points):
    x = np.linspace(x_min, x_max, num_points)
    for f in functions:
        y = f(x)#closure/high-order function     
        plt.plot(x, y)
    plt.gca().set_aspect("equal")
    #get current axis
    plt.show()
```

```python
plot_math_function([np.sin, lambda x: 2 * x + 3], -2 *np.pi, 2*np.pi, 5000)
```

```python
plot_math_function([lambda x: np.sqrt(25-x**2), lambda x: -np.sqrt(25-x**2)], -5, 5, 5000)
```
